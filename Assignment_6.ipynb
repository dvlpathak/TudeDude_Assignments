```python
from tkinter import *

def press(x):
    e.insert(END, x)

def clear():
    e.delete(0, END)

def equal():
    e.insert(END, "=" + str(eval(e.get())))

root = Tk()
root.title("Calculator")

e = Entry(root, width=20)
e.grid(row=0, column=0, columnspan=4)

for i, b in enumerate("789/456*123-0.+"):
    Button(root, text=b, width=4, command=lambda x=b: press(x)).grid(row=i//4+1, column=i%4)

Button(root, text="=", width=4, command=equal).grid(row=5, column=2)
Button(root, text="C", width=4, command=clear).grid(row=5, column=0)

root.mainloop()
```
